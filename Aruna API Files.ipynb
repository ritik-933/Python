```python

# Set up your Google Cloud Project ID, Dataset ID, and Table Name
project_id = 'eco-plating-407517'
dataset_id = 'ma_analytics'
credentials = service_account.Credentials.from_service_account_file('eco-plating-407517-8f535d4ac919.json',scopes=["https://www.googleapis.com/auth/cloud-platform"])


def createTableBQ(df,tableName):
    dataframe = df
    table_name = tableName

  # Define your BigQuery table schema


  # Initialize BigQuery client
    client = bigquery.Client(credentials=credentials, project=project_id)


  # Define the table reference
    table_ref = client.dataset(dataset_id).table(table_name)

  # Define the table object with schema
    table = bigquery.Table(table_ref)

  # Create the table in BigQuery
    client.create_table(table, exists_ok=True)  # This creates an empty table in BigQuery

  # Load DataFrame into the newly created table
    job = client.load_table_from_dataframe(dataframe, table_ref)
    job.result()  # Wait for job completion
    print("table create successfully,  +", table_name)
```
